# The effect of the garbage collector

Let's see whether the garbage collector makes any difference
in the speed with which text-fabric loads the BHSA. 

I work on a MacBook Pro (2021) with M1 Pro processor, 32 GB RAM. 

```python
from tf.app import use
```

# With GC

This is the default.

```python
A = use("ETCBC/bhsa:clone", checkout="clone", silent="verbose", withGc=True)
```

```python
A = use("ETCBC/bhsa:clone", checkout="clone", silent="verbose", withGc=True)
```

We see three time indications:

task | first time | second time | avg
--- | --- | --- | ---
init | 0.67 | 0.77 | 0.72
feats | 2.65 | 2.75 | 2.70
additional | 1.14 | 1.15 | 1.15




# Without GC

This is not how Text-Fabric runs by default.

```python
A = use("ETCBC/bhsa:clone", checkout="clone", silent="verbose", withGc=False)
```

```python
A = use("ETCBC/bhsa:clone", checkout="clone", silent="verbose", withGc=False)
```

We see three time indications:

task | first time | second time | avg
--- | --- | --- | ---
init | 1.18 | 1.26 | 1.22
feats | 3.13 | 3.23 | 3.18
additional | 1.38 | 1.15 | 1.27
