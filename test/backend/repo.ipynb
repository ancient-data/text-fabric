```python
import os
from gitlab import Gitlab, GitlabListError

from tf.advanced.repo import checkoutRepo
from tf.app import use
from tf.advanced.helpers import dm
```

```python
HOST = "gitlab.huc.knaw.nl"
ORG = "annotation"
REPO = "banks"
RELA = "tf"
VERSION = "0.2"
```

```python
def checkout(checkout, version=VERSION):
    dm("### On GitHub\n\n""")
    CH = checkoutRepo(
        host=None,
        org=ORG,
        repo=REPO,
        folder=RELA,
        version=version,
        checkout=checkout,
        withPaths=False,
    )
    
    dm("### On GitLab\n\n""")
    CL = checkoutRepo(
        host=HOST,
        org=ORG,
        repo=REPO,
        folder=RELA,
        version=version,
        checkout=checkout,
        withPaths=False,
    )
    dm("""\n---\n""")
    
def useTf(checkout, version=VERSION):
    dm("### On GitHub\n\n""")
    A = use(f"{ORG}/{REPO}:{checkout}", checkout=checkout, version=version, host=None)
    
    dm("### On GitLab\n\n""")
    A = use(f"{ORG}/{REPO}:{checkout}", checkout=checkout, version=version, host=HOST)
    dm("""\n---\n""")
```
